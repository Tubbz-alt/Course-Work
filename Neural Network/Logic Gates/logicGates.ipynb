```python
#Weights and Threshold Values for Gates

# AND Gate
ANDWeight = [ 0.5 , 0.5 ] 
ANDThreshold = 1

# NAND Gate
NANDWeight = [ -1 , -1 ]
NANDThreshold = -1

# OR Gate
ORWeight = [ 1 , 1 ]
ORThreshold = 1

# NOR Gate
NORWeight = [ -1 , -1 ]
NORThreshold = -0.5
# XOR Gate
    # is implemented using AND, OR and NAND
    
# NOT Gate
NOTWeight = -1
NOThreshold = 0


# Decide Function
def decideHelper( Weight , Threshold , Input ):
    sum_ = Weight[0]*Input[0] + Weight[1]*Input[1]
    if sum_ >= Threshold:
        return 1
    else:
        return 0
    
# Input Function    
def inputHelper():
    var = int(input())
    return var

# Print Function
def printHelper( inputSignal , outputSignal ):
    print("\n\tA\tB\t | \tC\n----------------------------------------\n")
    print("\t", inputSignal[0] , '\t' , inputSignal[1] , '\t | \t' , outputSignal , '\n' )

# Main Function
def main():
    while True:
        print("\n1. AND\n2. NAND\n3. OR\n4. NOR\n5. XOR\n6. NOT\n7. Exit")
        switchCase = int(input("\nEnter Your Choice\t:\t"))
        inputSignal = [ -1 , -1]
        if switchCase == 1:
            inputSignal[0] = inputHelper()
            inputSignal[1] = inputHelper()
            outputSignal = decideHelper( ANDWeight , ANDThreshold , inputSignal )
            printHelper( inputSignal , outputSignal )
            
        elif switchCase == 2:
            inputSignal[0] = inputHelper()
            inputSignal[1] = inputHelper()
            outputSignal = decideHelper( NANDWeight , NANDThreshold , inputSignal  )
            printHelper( inputSignal , outputSignal )
            
        elif switchCase == 3:
            inputSignal[0] = inputHelper()
            inputSignal[1] = inputHelper()
            outputSignal = decideHelper( ORWeight , ORThreshold , inputSignal )
            printHelper( inputSignal , outputSignal )
            
        elif switchCase == 4:
            inputSignal[0] = inputHelper()
            inputSignal[1] = inputHelper()
            outputSignal = decideHelper( NORWeight , NORThreshold , inputSignal )
            printHelper( inputSignal , outputSignal )
            
        elif switchCase == 5:
            inputSignal[0] = inputHelper()
            inputSignal[1] = inputHelper()
            finalInputSignal = []
            outputSignal1 = decideHelper( ORWeight , ORThreshold , inputSignal )
            outputSignal2 = decideHelper( NANDWeight , NANDThreshold , inputSignal  )
            finalInputSignal.append(outputSignal1)
            finalInputSignal.append(outputSignal2)
            outputSignal = decideHelper( ANDWeight , ANDThreshold , finalInputSignal )
            printHelper( inputSignal , outputSignal )
            
        elif switchCase == 6:
            inputSignal[0] = inputHelper()
            if NOTWeight*inputSignal[0] >= NOThreshold:
                outputSignal = 1

            else:
                outputSignal = 0
            print("\n\tA\t | \tC\n----------------------------------------\n")
            print("\t", inputSignal[0] , '\t | \t' , outputSignal , '\n' )
            continue
                
        elif switchCase == 7:
            break
        else:
            print("\nInvalid Choice\n")
            continue
        
# Calling Main Function
if __name__=='__main__':
    main()
```
